```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/laliga_player_stats_english.csv")
df.head(20)
```

```python
df.info()
```

```python
df.describe()
```

```python
# pie chart

df['Team'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python

# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['Team'])
```

```python
 Real_Madrid=df[df["Team"]=='Real Madrid']
 Real_Madrid.head(10)
```

```python
plt.figure(figsize=(16,8))
plt.title("Bar Plot of top scorers in Real Madrid")
plt.bar(list(Real_Madrid["Name"])[0:5],list(Real_Madrid["Games played"])[0:5])
```

```python
# Numberical Value # Histogram treat as categorical data distribution of data
plt.hist(df['Position'])
```

```python
# Distplot # probability density function --> The line is kde # skweness
sns.distplot(Real_Madrid['Games played'])
```

```python
# Boxplot # 5 number summary # median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier
sns.boxplot(Real_Madrid['Games played'])
```

```python
sns.scatterplot(x='Games started', y='Games played',hue='Corners',data=Real_Madrid)
plt.show()
```

```python
sns.lineplot(x='Interceptions',y='Recoveries',data=Real_Madrid)
```

```python
sns.heatmap(df.pivot_table(values='Interceptions',index='Team',columns='Tackles'))
```

```python
# Bar plot (Numerical - Categorical) # average age in pclass
sns.barplot(x='Position',y='Games played',hue='Goals scored',data=Real_Madrid)
```
